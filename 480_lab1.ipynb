```python
import numpy as np
from StringIO import StringIO
import matplotlib.pyplot as plt
from scipy.interpolate import interp1d
%matplotlib inline
from matplotlib.patches import Rectangle

```

```python
#location
lat = 48.4543
lon = -123.320
density = 2.7
waterdensity = 1.
drho = density-waterdensity

#import tide table data
tidedata = np.genfromtxt('tidedata.txt')
time = tidedata[:,0]
tide = tidedata[:,1]

#import experimental data
data = np.genfromtxt('data.txt')
d = data[:,0]
elevread = data[:,1]
t = data[:,2]
gread = data[:,3]
```

```python
# linear (local) interpolation over tide table times
f = interp1d(time,tide,kind='linear')

timeinterp = np.linspace(time[0],time[np.size(time)-1],1000)

#calculate tide corrections for measurements
tidecorr = f(t)
```

```python
#tide correction plot

fig = plt.figure(figsize=(6.5,3.5))
ax = fig.add_subplot(1,1,1)
ax.plot(time,tide,'.', color = 'k', label = "Tide Chart Values")
ax.plot(timeinterp,f(timeinterp),'k', alpha = 0.5, label = "Linear Interpolation")
ax.plot(t,tidecorr,'.', color = 'red',label = "Interpolated Corrections")
ax.set_xlabel("Time [min]"); ax.set_ylabel("Tide Correction [mGal]")
ax.legend(loc=0,fontsize='small')
fig.savefig('tideinterp.pdf')
```

```python
#conversion to mGal
gobs = 4611.03+ (gread-4500.00)*1.02611

#apply tide correction
gtc = gobs + tidecorr

#drift correction
index = np.size(t)-1
dgD = (gtc[index]-gtc[0])/(t[index]-t[0])*(t-t[0])

#lattitude correction
ds = d*np.tan(np.radians(26))/1000.
dgL = 0.8135*np.sin(2*np.radians(lat)*ds)

#free air correction
h = elevread[0]-elevread
dgFA = 0.3086*h

#Bouger Plate Correction
dgB = 0.04192*density*h

# Bouger Anomaly
gBase = gtc[0]
gB = gtc - gBase - dgD + dgL + dgFA - dgB

#output data summary
#for i in range(0,np.size(gB)-1):
#    print d[i],h[i], gtc[i], dgD[i], dgL[i], dgFA[i], dgB[i], gB[i]
```

```python
fig = plt.figure(figsize=(6.5,3.5))
ax = fig.add_subplot(1,1,1)
ax.plot(d[:index],gB[:index],color='red',alpha=0.3)
ax.plot(d[:index],gB[:index],'.',color='red')
ax.set_xlabel("Distance [m]")
ax.set_ylabel("Bouger Anomaly [mGal]")
fig.savefig('bougeranomaly.pdf')
```

```python
#linear interpolation of bouger anomaly funciton
bouger = interp1d(d[:index],gB[:index])
dinter = np.linspace(d[0],d[np.size(d)-2],10000)
bougerinter = bouger(dinter)

#maximum anomaly manually chosen at 30m based on knowledge of reservoir
xmax = 30.0
dgmax = bouger(xmax)

#function to find location of bouger anomaly value
def find_nearest(array,value):
    idx = (np.abs(array-value)).argmin()
    return dinter[idx]

#locations of half and quarter widths
x14 = find_nearest(bougerinter,dgmax/4.)
x12 = find_nearest(bougerinter,dgmax/2.)
x34 = find_nearest(bougerinter,3.*dgmax/4.)
#print x14,x12,x34

#depth to center of slab
zc1 = np.abs(x12-x34)
zc2 = np.abs(x12-x14)
zcc2 = np.abs(dgmax/2-dgmax/4)

print zc1,zc2
print "zc (averaged):"
print (zc1+zc2)/2.

#slab thickness
G = 0.04192/(2*np.pi)
h = dgmax/(2*G*np.pi*drho)
print h

```

```python
fig = plt.figure(figsize=(6.5,3.5))
ax = fig.add_subplot(1,1,1)
ax.plot(dinter,bougerinter,color='red',alpha=0.5,lw=1.5)
ax.plot(d[:index],gB[:index],'.',color='red')

ax.plot((0,60),(dgmax,dgmax),'k--')
ax.plot((0,60),(dgmax/2,dgmax/2),color='blue',ls='--')
ax.plot((0,60),(dgmax/4,dgmax/4),color='green',ls='--')
ax.plot((0,60),(3*dgmax/4,3*dgmax/4),color='purple',ls='--')

ax.plot((x34,x34),(0,-0.45),'purple')
ax.plot((x12,x12),(0,-0.45),'green')
ax.plot((x14,x14),(0,-0.45),'blue')

ax.annotate("$\Delta g_{max}$",(30,-0.39))
ax.annotate("$0.25 \Delta g_{max}}$",(30,-0.115))
ax.annotate("$0.5 \Delta g_{max}$",(30,-0.21))
ax.annotate("$0.75 \Delta g_{max}$",(30,-0.30))

ax.set_xlabel("Distance [m]")
ax.set_ylabel("Bouger Anomaly [mGal]")
ax.set_xlim(0,35)
fig.savefig('bougeranomalyannotate.pdf')
```
